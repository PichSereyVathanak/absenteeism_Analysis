### Part 1: Data Preprocessing

```python
#import library and ignore warning 
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
#don't set the limited of display rows and columns of pd
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
#load the dataset from csv file
raw_csv_data = pd.read_csv('../data/raw/Absenteeism_data.csv')

#copy the data into dataframe
df = raw_csv_data.copy()

#read the header of file
df.head(10)
```

```python
#lowercase & replace white space with underscore(_) for the columns name
df.columns = (df.columns
    .str.strip() #remove white space from head & tail
    .str.replace(' ', '_') #replace whitespace with _
    .str.lower()) #lowercase the columns name
df.head()
```

```python
#check number of rows and columns
df.shape
```

```python
#describe the dataset
df.describe(include='all')
```

```python
#check datatype
df.info()
```

```python
#drop the ID column
df = df.drop(columns='id', axis=1)
# df.drop(['ID], axis =1)
df.head()
```

```python
#min max and unique values of Reason for Absence

print(f'min: {df['reason_for_absence'].min()}')
print(f'max: {df['reason_for_absence'].max()}')
print(f'Unique Values: {df['reason_for_absence'].unique()}')
print(f'Sorted: {sorted(df['reason_for_absence'].unique())}')
```

```python
#get dummies()

reason_columns = pd.get_dummies(df['reason_for_absence'], dtype= int)
reason_columns.head()
```

```python
#sum each row in new columns call check
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns.head()
```

```python
#sum values in check column
reason_columns['check'].sum(axis=0)
```

```python
#verify that all values in column check is 1
reason_columns['check'].unique()
```

```python
#drop the check column
reason_columns = reason_columns.drop(['check'], axis = 1)
reason_columns.head()
```

```python
reason_columns.head()
```

```python
#drop the first column which is number 0: unknwon reason of absense
reason_columns = pd.get_dummies(df['reason_for_absence'], drop_first=True, dtype=int)
reason_columns.head()
```

```python
#group the reason into same key characteristics
reason_columns.columns.values
```

- Group 1: related to disease 1-14
- Group 2: related to pregnancy and give birth 15-17
- Group 3: poisoning or sign not elsewhere categorized 18-21
- Group 4: ligh reason of absenss 22-28

```python
#classified the data
#loc[] is label-based [1, 2] 
#1: number of rows
#2: columns from the start to end
reason_columns.loc[:, '1':'14']
```

```python
#iloc using index base which mean that start from 0 to the len -1 of our dataset
reason_columns.iloc[:, 0:14]
```

```python
#group the reason into 4 columns base on our defined class
#axis = 1 indidcate find max for each row
# loc use the labal(columns name)
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
#check
reason_type_1
```

```python
df.head()
```

```python
#drop reason_for_absense
df = df.drop(['reason_for_absence'], axis=1)
df.head()
```

```python
#concatenate the 4 reson into our main df
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis =1)
df.head()
```

```python
#rename the 4 reason group of absense
#numerical columns name so no ' '
df.rename(columns={
    0: 'reason_1',
    1: 'reason_2',
    2: 'reason_3',
    3: 'reason_4'}, inplace=True)
```

```python
df.head()
```

```python
#re-order the column
df.columns.values
```

```python
reorder_columns = ['reason_1', 'reason_2', 'reason_3', 'reason_4', 
                   'date', 'transportation_expense', 'distance_to_work', 'age',
                   'daily_work_load_average', 'body_mass_index', 'education',
                   'children', 'pets', 'absenteeism_time_in_hours']

df = df[reorder_columns]

df.head()
```
