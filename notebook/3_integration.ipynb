```python
cols = df_new_obs.columns.tolist()
insert_query = f"INSERT INTO predicted_outputs ({', '.join(cols)}) VALUES "

values_parts = []
for i in range(df_new_obs.shape[0]):
    row = df_new_obs.iloc[i]
    row_items = []
    for val in row:
        if pd.isna(val):
            row_items.append("NULL")
        elif isinstance(val, str):
            row_items.append("'" + val.replace("'", "''") + "'")
        else:
            row_items.append(str(val))
    values_parts.append("(" + ", ".join(row_items) + ")")

insert_query += ", ".join(values_parts) + ";"
cursor.execute(insert_query)
connection.commit()
```
